```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv('recursos.csv', parse_dates=['Fecha'])

# Graficar el consumo de memoria y CPU
plt.figure(figsize=(10, 5))
plt.plot(df['Fecha'], df['Memoria(%)'], label='Memoria', color='blue')
plt.plot(df['Fecha'], df['CPU(%)'], label='CPU', color='red')

# Aadir etiquetas y leyenda
plt.title('Consumo de Memoria y CPU')
plt.xlabel('Tiempo')
plt.ylabel('Porcentaje')
plt.legend()

# Rotar las etiquetas del eje x para una mejor visualizacin
plt.xticks(rotation=45)

# Mostrar la grfica
plt.grid(True)
plt.tight_layout()
plt.show()
```
